#### <font color='red'> Definição 1 </font>

<font color='red'> Seja $G$ um grafo simples. Uma **árvore geradora** de $G$ é um subgrafo de $G$ que é uma árvore contendo todos os vértices de $G$.

#### <font color='blue'> Exemplo 1 </font>

<font color='blue'> Encontre uma árvore geradora do grafo simples $G$ mostrado na figura a seguir:<br><br>
 <center> <img src="fig18_2.jpg" alt="jpg_python" width="280" height=""> </center>  

##### <font color='blue'> Solução </font>

<font color='blue'> O grafo $G$ está conectado, mas não é uma árvore, pois contém circuitos simples. Remova as arestas $\{d,c\}$ e $\{d,e\}$. Isso elimina um circuito simples, e o subgrafo resultante ainda está conectado e ainda contém todos os vértices de $G$. Em seguida, remova a aresta $\{e, f\}$ para eliminar um segundo circuito simples. Por fim, remova a aresta $\{g,h\}$ para produzir um grafo simples sem circuitos simples. Este subgrafo é uma árvore geradora, pois é uma árvore que contém todos os vértices de $G$. A sequência de remoções de arestas usada para produzir a árvore de abrangência é ilustrada a seguir:<br><br>

#### <font color='red'> Teorema 1 </font>

<font color='red'> Um grafo simples é conectado se, e somente se, tiver uma árvore geradora.

##### <font color='red'> Demonstração </font>

<font color='red'> Primeiro, suponha que um grafo simples tenha uma árvore geradora $T$. $T$ contém todos os vértices de $G$. Além disso, existe um caminho em $T$ entre quaisquer dois dos seus vértices. Como $T$ é um subgrafo de $G$, existe um caminho em $G$ entre quaisquer dois de seus vértices. Portanto, $G$ está conectado.<br><br>
 Agora suponha que $G$ esteja conectado. Se não for uma árvore, ela deve conter um circuito simples. Remova uma aresta de um desses circuitos simples. O subgrafo resultante tem uma aresta a menos, mas ainda contém todos os vértices de $G$ e está conectado. Esse subgrafo ainda está conectado pois, quando dois vértices estão conectados por um caminho contendo a aresta removida e esse caminho faz parte de um circuito simples que retorna para algum desses vértices, então eles são conectados por um caminho que não contém essa aresta. Podemos construir esse caminho inserindo-o no caminho original, no ponto em que a aresta removida era uma vez, o circuito simples com essa aresta removida. Se este subgrafo não é uma árvore, tem um circuito simples; assim como antes, remova uma aresta que esteja em um circuito simples. Repita este processo até que nenhum circuito simples permaneça. Isso é possível pois há apenas um número finito de arestas no grafo. O processo termina quando não há circuitos simples. Uma árvore é produzida pois o grafo permanece conectado conforme as arestas são removidas. Esta árvore é uma árvore geradora pois contém todos os vértices de $G$.

<font color='black'><span style="font-size: 22px;"><strong>18.2 Busca em Profundidade</strong></span></font>

<hr style="border: 2px solid black;">
</div>

 A demonstração do **Teorema 1** fornece um algoritmo para encontrar árvores geradoras, removendo arestas de circuitos simples. Esse algoritmo é ineficiente, pois requer que circuitos simples sejam identificados. Em vez de construir árvores geradoras removendo arestas, podemos construí-las adicionando arestas sucessivamente por meio da **busca em profundidade**. Nós formaremos uma árvore enraizada, e a árvore geradora será o grafo não orientado subjacente a esta árvore enraizada. 

<font color='red'> Arbitrariamente, escolha um vértice do grafo como a raiz. Forme um caminho começando neste vértice, adicionando sucessivamente vértices e arestas, onde cada nova aresta incide com o último vértice no caminho e um vértice que ainda não esteja no caminho. Continue adicionando vértices e arestas a esse caminho o maior tempo possível. Se o caminho percorrer todos os vértices do grafo, a árvore que consiste nesse caminho é uma árvore de abrangência. No entanto, se o caminho não passar por todos os vértices, mais vértices e arestas deverão ser adicionados. Volte para o próximo ao último vértice no caminho e, se possível, forme um novo caminho começando neste vértice passando por vértices que ainda não foram visitados. Se isso não puder ser feito, mova de volta outro vértice no caminho, ou seja, dois vértices de volta no caminho e tente novamente. Repita este procedimento, começando no último vértice visitado, movendo-se de volta pelo caminho um vértice de cada vez, formando novos caminhos que sejam o maior tempo possível até que não seja possível adicionar mais arestas. Como o grafo tem um número finito de arestas e está conectado, esse processo termina com a produção de uma árvore geradora. Cada vértice que termina um caminho em um estágio do algoritmo será uma folha na árvore enraizada, e cada vértice onde um caminho é construído a partir deste vértice será um vértice interno.</fotn>

<font color='blue'> Use a busca em profundidade para encontrar uma árvore geradora para o grafo $G$ mostrado na figura a seguir:<br><br>

##### <font color='blue'> Solução </font>

<font color='blue'> Os passos usados pela primeira busca em profundidade para produzir uma árvore geradora de $G$ são mostrados na figura a seguir. Nós começamos arbitrariamente com o vértice $a$ (mas pode ser outro). Um caminho é construído adicionando sucessivamente as arestas (*seguindo alguma ordem, digamos a lexicográfica*) incidentes com os vértices que ainda não estão no caminho, desde que isso seja possível. Isso produz um caminho $a,c,b$ (*note que outros caminhos poderiam ter sido construídos caso usássemos um critério diferente da ordem lexicográfica*). Em seguida, volte para $c$. Agora adicionamos o caminho $d,e,f$. Voltando para o vértice $e$, adicionamos o caminho $g,h,i$. Voltando para o vértice $h$, adicionamos, por último, o vértice $j$. Temos assim uma árvore geradora de $G$ obtida por meio da busca em profundidade:<br><br>
 <center> <img src="fig18_6.jpg" alt="jpg_python" width="800" height=""> </center>

As arestas selecionadas pela busca em profundidade de um grafo são chamadas **arestas da árvore**. Todas as outras arestas do grafo devem conectar um vértice a um ancestral ou descendente desse vértice na árvore. Essas arestas são chamadas de \underline{arestas de retorno}. No **Exemplo 2** temos que as arestas de retorno são dadas por $\{d,f\}$ e $\{e,h\}$

#### <font color='green'> Palavras dos Mestres </font>

<font color='green'> No **Exemplo 2** poderíamos ter optado pela escolha dos seguintes caminhos $a,c,d,f,e,g,h,j$ (caminho longo) voltando em seguida para o vértice $h$ acrescentando a aresta $\{h,i\}$, depois retornando ao vértice $c$ e acrescentando a aresta $\{b,c\}$. A princípio não parece haver nenhum problema com esta escolha. Contudo, em uma situação prática, ao darmos nomes aos vértices (no nosso caso, nomes formados por uma única letra) podemos usar isso em nosso favor, principalmente quando precisarmos indicar à máquina qual escolha tomar. Note que temos uma capacidade, pelo menos neste caso, de enxergar um caminho mais longo quase que de imediato, muito diferente de uma máquina com uma programação simplificada, sem recursos visuais! Lembre-se de que você está estudando esta matéria para usá-la em seu curso, futuramente implementando-a e não somente para passar na prova.

#### <font color='darkblue'> Algoritmo:  Busca em Profundidade

<font color='darkblue'> procedure $DFS$ ($G$: grafo conexo com vértices $v_{1}, v_{2} \ldots, v_{n}$)<br><br>
 $ \ \ \ $ $T:=$ árvore que consiste apenas do vértice $v_{1}$<br><br>
 $ \ \ \ $ *visit*($v_{1}$)<br><br>
 $ \ \ \ $ *procedure*<br><br>
 $ \ \ \ $ *for* cada vértice $w$ adjacente a $v$ e ainda não em $T$<br><br>
 $ \ \ \ $ *begin*<br><br>
 $ \ \ \ \ \ \ $ adicione o vértice $w$ a aresta $\{v.w\}$ a $T$<br><br>
 $ \ \ \ \ \ \ $ *visit*($w$)<br><br>
 $ \ \ \ $ *end*<br><br>

<hr style="border: 2px solid black;">

Também podemos produzir uma árvore geradora de um grafo simples pelo uso da **busca em largura**. Novamente, uma árvore enraizada será construída e o grafo não direcionado subjacente dessa árvore enraizada forma a árvore geradora. 

<font color='red'> Arbitrariamente, escolha uma raiz nos vértices do grafo. Em seguida, adicione todas as arestas incidentes neste vértice. Os novos vértices adicionados neste estágio se tornam os vértices no nível $1$ na árvore geradora. Em seguida, para cada vértice no nível $1$, visitado em ordem, adicione cada aresta incidente a este vértice na árvore, desde que ele não produza um circuito simples. Ordene arbitrariamente os filhos de cada vértice no nível $1$. Isso produz os vértices no nível $2$ da árvore. Siga o mesmo procedimento até que todos os vértices da árvore tenham sido adicionados. O procedimento termina porque há apenas um número finito de arestas no grafo. Uma árvore geradora é produzida pois produzimos uma árvore contendo todos os vértices do grafo. Um exemplo de pesquisa em largura é dado no **Exemplo 3**. *Novamente, devemos usar alguma ordenação para a escolha dos objetos.*

<font color='blue'> Use a busca em largura para encontrar uma árvore geradora para o grafo mostrado na figura a seguir:<br><br>
 <center> <img src="fig18_7.jpg" alt="jpg_python" width="320" height=""> </center>


##### <font color='blue'> Solução </font>

<font color='blue'> Os passos do primeiro procedimento de busca são mostrados na figura a seguir. Nós escolhemos o vértice $a$ (vértice do nível $0$) como a raiz. Em seguida, adicionamos as arestas incidentes com todos os vértices adjacentes a $a$. Assim, adicionamos os vértices (e suas arestas) $b$ e $d$ (vértices do nível $1$). A partir de $b$ adicionamos os vértices $c$ e $e$ (vértices do nível $2$). Ainda no nível $1$, a partir de $d$ adicionamos o vértice $h$ (vértice do nível $2$). A partir de $c$ adicionamos o vértice $f$ (nível $3$). A partir de $e$ adicionamos o vértice $i$ (nível $3$). A partir de $h$ adicionamos $k$ (nível $3$). A partir de $f$ adicionamos $g$ e $j$ (nível $4$). A partir de $h$ adicionamos $k$. Note que não adicionamos mais vértice algum a partir de $i$. Do contrário, teríamos um circuito simples. A partir de $g$ adicionamos $l$ (nível $5$) e, por fim, a partir de $k$ adicionamos $m$. Temos assim uma árvore geradora de $G$ obtida por meio da busca em largura.<br><br>
 <center> <img src="fig18_8.jpg" alt="jpg_python" width="800" height=""> </center>

#### <font color='darkblue'> Algoritmo:  Busca em Largura

<font color='darkblue'>procedure $BFS$ ($G$: grafo conexo com vértices $v_{1}, v_{2} \ldots, v_{n}$)<br><br>
 $\ \ \ $ $T:=$ árvore que consiste apenas do vértice $v_{1}$<br><br>
 $\ \ \ $ $L:=$ lista vazia<br><br>
 $\ \ \ $ ponha $v_{1}$ na lista $L$ do vértices não processados<br><br>
 $\ \ \ $ **while** $L$ não estiver vazia<br><br>
 $\ \ \ $ **begin**<br><br>
 $\ \ \ \ \ \ $ remova o primeiro vértice, $v$, de $L$.<br><br>
 $\ \ \ \ \ \ $ **for** cada vizinho $w$ de $v$<br><br>
 $\ \ \ \ \ \ \ \ \ $ **if** $w$ não estiver em $L$ e não estiver em $T$ **then**<br><br>
 $\ \ \ \ \ \ \ \ \ $ **begin**<br><br>
 $\ \ \ \ \ \ \ \ \ $ adicione $w$ ao final da lista $L$<br><br>
 $\ \ \ \ \ \ \ \ \ $ adicione $w$ e as arestas $\{v,w\}$ a $T$<br><br>
 $\ \ \ \ \ \ $ **end**<br><br>
 $\ \ \ $ **end**<br><br>

Para os exemplos a seguir utilizamos as árvores dadas nos **Exemplos 1**, **2** e **3**.

```python
H = [('a', 'b'), ('a', 'd'), ('b', 'c'), ('c', 'd'), ('c', 'e'), ('c', 'f'),\
     ('d', 'e'), ('e', 'f'), ('f', 'g'), ('f', 'h'), ('g', 'h')]

I = [('a', 'c'), ('b', 'c'), ('c', 'd'), ('d', 'e'), ('d', 'f'), ('e', 'f'),\
     ('e', 'g'), ('e', 'h'), ('g', 'h'), ('h', 'i'), ('h', 'j')]

J = [('a', 'b'), ('a', 'd'), ('b', 'c'), ('b', 'e'), ('c', 'f'), ('d', 'e'),\
     ('d', 'h'), ('e', 'f'), ('e', 'i'), ('f', 'g'), ('f', 'j'), ('g', 'j'),\
     ('g', 'l'), ('h', 'i'), ('h', 'k'), ('i', 'j'), ('i', 'k'), ('k', 'm')]
```

#### Busca em Profundidade

```python
import networkx as nx

# Cria um grafo e adiciona as arestas
G = nx.Graph()
G.add_edges_from(H) # Troque H por I e J

# Realiza a busca em profundidade a partir do vértice 'a'
dfs_edges = list(nx.dfs_edges(G, 'a'))
dfs_nodes = list(nx.dfs_preorder_nodes(G, 'a'))

# Imprimindo arestas e vértices da DFS
print(f'Arestas da DFS: {dfs_edges}')
print(f'Vertices da DFS: {dfs_nodes}')
```

Podemos aproveitar o código acima para desenhar a árvore geradora resultante da busca em profundidade:

```python
import matplotlib.pyplot as plt

# Cria uma nova instância de grafo para a árvore geradora DFS
dfs_tree = nx.Graph()

# Adiciona as arestas à árvore geradora DFS
dfs_tree.add_edges_from(dfs_edges)

# Desenha a árvore geradora DFS
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=800, font_size=15, pos=nx.spring_layout(G))
plt.title("Grafo Original com Arestas da DFS destacadas")
plt.show()

# Desenhando também a árvore geradora DFS
plt.figure(figsize=(12, 8))
nx.draw(dfs_tree, with_labels=True, node_color='lightblue', edge_color='red', node_size=800, font_size=15, pos=nx.spring_layout(dfs_tree))
plt.title("Árvore Geradora - Busca em Profundidade")
plt.show()
```

#### Busca em Largura

```python
import networkx as nx

# Cria um grafo e adiciona as arestas
G = nx.Graph()
G.add_edges_from(H) # Troque H por I e J

# Realiza a busca em largura a partir do vértice 'a'
bfs_order_edges = list(nx.bfs_edges(G, 'a'))
bfs_order_nodes = ['a'] + [v for u, v in bfs_order_edges]

print(f'Arestas: {bfs_order_edges}')
print(f'Vértices: {bfs_order_nodes}')
```

Podemos aproveitar o código acima para desenhar a árvore geradora resultante da busca em largura:

```python
import matplotlib.pyplot as plt

# Cria uma nova instância de grafo para a árvore geradora BFS
bfs_tree = nx.Graph()
bfs_tree.add_edges_from(bfs_order_edges)

# Desenha a árvore geradora BFS
plt.figure(figsize=(8, 6))
nx.draw(bfs_tree, with_labels=True, node_color='lightblue', edge_color='gray', node_size=800, font_size=15)
plt.title("BFS Spanning Tree")
plt.show()
```
