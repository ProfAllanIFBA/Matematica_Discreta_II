# <div style="margin-top: -20px;">
<hr style="border: 2px solid black;">

<font color='black'><span style="font-size: 22px;"><strong>16.1 Sistema de Endereçamento Universal</strong></span></font>

<hr style="border: 2px solid black;">
</div>

Procedimentos para atravessar todos os vértices de uma árvore ordenada com raízes dependem das ordenações dos filhos. Em árvores enraizadas ordenadas, os filhos de um vértice interno são mostrados da esquerda para a direita nos desenhos que representam esses grafos direcionados. Descreveremos uma maneira pela qual podemos ordenar totalmente os vértices de uma árvore enraizada ordenada. Para produzir esta ordenação, devemos primeiro rotular todos os vértices. Fazemos isso recursivamente:

#### Processo de organização

<font color='red'>1. Rotule a raiz com o inteiro $0$. Em seguida, rotule seus $k$ filhos (no nível $1$) da esquerda para a direita com $1, 2, 3, \ldots, k$.<br><br>
 2. Para cada vértice $v$ no nível $n$ com o rótulo $A$, rotule seus filhos $k_{v}$, como eles são extraídos da esquerda para a direita, com $A.1, A.2, \ldots, A.k_{v}$.<br><br>
 Seguindo este procedimento, um vértice $v$ no nível $n$, para $n \geq 1$, é rotulado $x_{1}.x_{2}. \ldots .x_{n}$, onde o caminho único da raiz para $v$ passa pelo vértice $x_{1}st$ no nível $1$, o vértice $x_{2}$ no nível $2$ e assim por diante. Essa rotulagem é chamada de sistema de endereço universal da árvore ordenada.

Podemos ordenar totalmente os vértices usando o ordenamento lexicográfico de seus rótulos no sistema de endereço universal. O vértice rotulado $x_{1}.x_{2}. \ldots .x_{n}$ é menor que o vértice identificado como $y_{1}.y_{2}. \ldots .y_{m}$ se houver um, $0 \leq i \leq n$, com $x_{1} = y_{1}, x_{2} = y_{2},. . . x_{i-1} = y_{i-1}$ e $x_{i} < y_{i}$; ou se $n < m$ e $x_{i} = y_{i}$ para $i = 1, 2, \ldots n.$

<font color='black'><span style="font-size: 22px;"><strong>16.2 Algoritmos de Travessia</strong></span></font>

<hr style="border: 2px solid black;">
</div>

Os procedimentos para visitar sistematicamente todos os vértices de uma árvore com raiz ordenada são chamados de algoritmos de travessia. Descreveremos três dos algoritmos mais comumente usados, a **travessia em pré-ordem**, a **travessia ordem simétrica** e a **travessia pós-ordem**. Cada um desses algoritmos pode ser definido recursivamente. Primeiro definimos a travessia da pré-ordem.

#### <font color='red'> Definição 1: Travessia em Pré-Ordem </font>

<font color='red'> Seja $T$ uma árvore enraizada ordenada com raiz $r$. Se $T$ consiste apenas de $r$, então $r$ é o percurso de pré-ordem de $T$. Caso contrário, suponha que $T_{1}, T_{2}, \ldots , T_{n}$ são as subárvores em $r$ da esquerda para a direita em $T$. A travessia da pré-ordem começa visitando $r$. Ele continua percorrendo $T_{1}$ na pré-ordem, depois $T_{2}$ na pré-ordem, e assim por diante, até que $T_{n}$ seja atravessado na pré-ordem.

#### <font color='blue'> Exemplo 2 </font>

<font color='blue'> Determine a sequência de vértices a serem percorridos em uma travessia em pré-ordem da árvore enraizada $T$ dada na figura a seguir:<center> <img src="fig16_2.jpg" alt="jpg_python" width="350" height=""> </center>

#### <font color='blue'> Solução

<font color='blue'> Primeiro pesquisamos a raiz $a$. Note que $a$ tem três subárvores de raízes $b$, $c$ e $d$. Consideremos a subárvore mais a esquerda, isto é a subárvore de raiz $b$. Pesquisemos a raiz $b$ e sua subárvore esquerda de raiz $e$ (e em seguida seus filhos se existissem). Prosseguindo, pesquisamos a subárvore de raiz $f$ começando por esta raiz. Note que a subárvore de raiz $f$ tem filhos $j$ e $k$. Pesquisamos o filho $j$ e seus filhos $l$ e $m$. Tendo pesquisado toda a subárvore de raiz $j$ partimos para o vértice $k$ donde finalizamos a travessia da subárvore de raiz $b$. Como $c$ é o irmão direito mais próximo de $b$ partiremos para a subárvore de raiz $c$ pesquisando este vértice inicialmente e seus filhos $g$ e $h$. Tendo atravessado toda a subárvore de raiz $c$ resta-nos apenas a subárvore de raiz $d$. Por fim, pesquisamos o vértice $d$ e, em seguida, ao chegar no vértice $i$, completamos a travessia da árvore $T$ em pré-ordem. Assim, temos a seguinte ordenação dos vértices em uma travessia em pré-ordem: $a,b,e,f,j,l,m,k,c,g,h,d,i$.

#### <font color='darkgreen'> Truqe para a travessia em pré-ordem </font>

<font color='darkgreen'> Há um truque simples para conferirmos se a ordem dos vértices de uma árvore em uma travessia em pré-ordem está correta. Para tando, comece fazendo um pequeno sinal(símbolo, marca etc) à esquerda de cada vértice na árvore em questão. Depois, partindo da esquerda da raiz ligue todos estes sinais em uma linha contínua "circulando" a árvore nas proximidades do seu vértice. A ordem com que os sinais foram sendo pintados corresponde à pesquisa em pré-ordem. Esta regra não lhe isenta de saber o processo de travessia uma vez que se quer entender o mecanismo de travessia para uma posterior implementação. #### <font color='red'> Definição 2: Travessia em Ordem Simétrica </font>

<font color='red'> Seja $T$ uma árvore enraizada ordenada com raiz $r$. Se $T$ consiste apenas de $r$, então $r$ é a travessia ordem simétrica de $T$. Caso contrário, suponha que $T_{1}, T_{2}, \ldots , T_{n}$ são as subárvores da esquerda para a direita. A travessia ordem simétrica começa atravessando $T_{1}$ e em seguida visitando $r$. Ele continua percorrendo $T_{2}$ em ordem simétrica, depois $T_{3}$ em ordem simétrica, $\ldots$ e finalmente $T_{n}$ em ordem simétrica.

#### <font color='blue'> Exemplo 3 </font>

<font color='blue'> Determine a sequência de vértices a serem percorridos em uma travessia em ordem simétrica da árvore enraizada $T$ dada no **Exemplo 2**:<center> <img src="fig16_2.jpg" alt="jpg_python" width="350" height=""> </center>

##### <font color='blue'> Solução

<font color='blue'> Note que a raiz $a$ de $T$ tem subárvores $T_{1}$ de raiz $b$, $T_{2}$ de raiz $c$ e $T_{3}$ de raiz $d$. Antes de pesquisamos a raiz $a$ devemos pesquisar a subárvore $T_{1}$. Mas a subárvore $T_{1}$ tem ela duas subárvores $T^{b}_{1}$ de raiz $e$ e $T^{b}_{2}$ de raiz $f$. Neste caso, pesquisamos primeiramente o vértice $e$ (pois não possui mais subárvores) e em seguida sua raiz $b$. Partimos então para a subárvore da direita $T^{b}_{2}$. Mas esta tem outras duas subárvores $T^{f}_{1}$ de raiz $j$ e $T^{f}_{2}$ de raiz $k$. Assim, pesquisamos em $T^{f}_{1}$ o vértice $l$(pois não possui mais subárvores), em seguida a raiz $j$ de $T^{f}_{1}$ e por último o vértice $m$. Tendo feito a travessia de $T^{f}_{1}$ partimos para a raiz $f$ e em seguida para o vértice $k$ da árvore $T^{b}_{2}$. Como atravessamos toda subárvore $T_{1}$ de raiz $b$ pesquisamos a raiz $a$ e em seguida partimos para a subárvore $T_{2}$. Mas $T_{2}$ tem duas subárvores $T^{c}_{1}$ de raiz $g$ e $T^{c}_{2}$ de raiz $h$. Neste caso, pesquisamos o vértice $g$ (sem filhos), a raiz $c$ e por último o vértice $h$. Tendo atravessado toda subárvore $T_{2}$ partimos para a subárvore $T_{3}$. Neste caso, pesquisamos o vértice $i$ e depois a raiz $d$ completando a travessia da árvore $T$ em ordem simétrica. Assim, temos a seguinte ordenação dos vértices em uma travessia em ordem simétrica: $e,b,l,j,m,f,k,a,g,c,h,i,d$.

#### <font color='darkgreen'> Truqe para a travessia em ordem simétrica </font>

<font color='darkgreen'> Mais uma boa notícia. Há, também, um truque simples para conferirmos se a ordem dos vértices de uma árvore em uma travessia em ordem simétrica está correta. Para tando, comece fazendo um pequeno sinal(símbolo, marca etc) abaixo de cada vértice (entre a primeira aresta e as demais caso existam abaixo do vértice) na árvore em questão. Depois, partindo da esquerda da raiz ligue todos estes sinais em uma linha contínua "circulando" a árvore nas proximidades do seu vértice. A ordem com que os sinais foram sendo pintados corresponde à pesquisa em ordem simétrica. #### <font color='red'> Definição 3: Travessia em Pós-Ordem </font>

<font color='red'> Seja $T$ uma árvore enraizada ordenada com raiz $r$. Se $T$ consiste apenas em $r$, então $r$ é a travessia de pós-ordem de $T$. Caso contrário, suponha que $T_{1}, T_{2}, \ldots , T_{n}$ são as subárvores da esquerda para a direita. A travessia de pós-ordem começa atravessando $T_{1}$ no pós-ordem, depois $T_{2}$ no pós-ordem, $\ldots$ , em seguida, $T_{n}$ no pós-ordem e termina visitando $r$.

#### <font color='blue'> Exemplo 4 </font>

<font color='blue'> Determine a sequência de vértices a serem percorridos em uma travessia em pós-ordem da árvore enraizada $T$ dada no **Exemplo 2**:<center> <img src="fig16_2.jpg" alt="jpg_python" width="350" height=""> </center>

##### <font color='blue'> Solução

<font color='blue'> Note que a raiz $a$ tem três subárvores $T_{1}$ de raiz $b$, $T_{2}$ de raiz $c$ e $T_{3}$ de raiz $d$. A subárvore $T_{1}$ tem subárvores $T^{b}_{1}$ de raiz $e$ e $T^{b}_{2}$ de raiz $f$. Neste caso pesquisamos o vértice $e$ e partimos para $T^{b}_{2}$. Note que $T^{b}_{2}$ tem duas subárvores $T^{f}_{1}$ de raiz $j$ e $T^{f}_{2}$ de raiz $k$. Assim, partimos para a subárvore $T^{f}_{1}$ que possui duas subárvores $T^{j}_{1}$ de raiz $l$ e $T^{j}_{2}$ de raiz $m$. Pesquisamos então o vértice $l$ (nosso segundo vértice é pesquisado somente agora!), em seguida o vértice $m$ e só então a raiz $j$ de $T^{f}_{1}$. Agora pesquisamos o vértice $k$ correspondente à subárvore $T^{f}_{2}$ para então pesquisar o vértice $f$ finalizando a travessia da subárvore $T^{b}_{2}$. Tendo finalizado a subárvore $T^{b}_{2}$ pesquisamos o vértice $b$ finalizando $T_{1}$. Agora, passamos para a subárvore $T_{2}$ a qual tem duas subárvores $T^{c}_{1}$ de raiz $g$ e $T^{c}_{2}$ de raiz $h$. Neste caso, pesquisamos o vértice $g$, em seguida o vértice $h$ e por último o vértice $c$ finalizando a travessia de $T_{2}$. Agora podemos partir para a subárvore $T_{3}$. Note que esta tem somente uma subárvore $T^{d}_{1}$ de raiz $i$. Sendo assim, pesquisamos o vértice $i$ e por último o vértice $d$ finalizando $T_{3}$. Como fizemos a travessia de $T_{1}$, $T_{2}$ e $T_{3}$ em pós-ordem podemos percorrer, finalmente, a raiz $a$ completando a travessia em pós-ordem da árvore $T$. Assim, temos a seguinte ordenação dos vértices em uma travessia em pós-ordem: $e,l,m,j,k,f,b,g,h,c,i,d,a$.

#### <font color='darkgreen'> Truqe para a travessia em pós-ordem </font>

<font color='darkgreen'> Analogamente aos processos de travessia em pré-ordem e também em ordem simétrica há, também, um truque simples para conferirmos se a ordem dos vértices de uma árvore em uma travessia em pós-ordem está correta. Para tando, comece fazendo um pequeno sinal(símbolo, marca etc) à direita de cada vértice na árvore em questão. Depois, partindo da esquerda da raiz ligue todos estes sinais em uma linha contínua "circulando" a árvore nas proximidades do seu vértice. A ordem com que os sinais foram sendo pintados corresponde à pesquisa em pós-ordem.
```python
arvore = [('a', 'b'), ('a', 'c'), ('a', 'd'), ('b', 'e'), ('b', 'f'), ('c', 'g'),\
          ('c', 'h'), ('d', 'i'), ('f', 'j'), ('f', 'k'), ('j', 'l'), ('j', 'm')]
```

#### Travessia em pré-ordem

```python
# Importa as bibliotecas necessárias
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import warnings  # Módulo referente a avisos

# Cria o grafo
G = nx.DiGraph()  # Note que estamos usando DiGraph (grafo direcionado) agora

# Adicione arestas ao grafo
G.add_edges_from(arvore)

# Define a função de travessia em pré-ordem
def preorder_traversal(G, node, visited=[]):
    # Adiciona o nó atual à lista de visitados
    visited.append(node)

    # Para cada vizinho do nó atual
    for neighbor in G.neighbors(node):
        # Se o vizinho ainda não foi adicionado à lista de visitados
        if neighbor not in visited:
            # Realiza a travessia em pré-ordem no vizinho
            preorder_traversal(G, neighbor, visited)

    return visited

# Executa a travessia em pré-ordem a partir do nó 1
visited = preorder_traversal(G, arvore[0][0]) # Especifique outro nó caso a raiz não seja
                                              # o primeiro vertice lido

warnings.simplefilter("ignore") # Omite alguns avisos que não afetam a execução

# Exibe a lista de nós visitados
print(visited)

# PARA DESENHAR O GRAFO
# Desenha o grafo utilizando layout hierárquico
pos = nx.drawing.nx_pydot.graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, arrows=False)

# Exibe a figura
plt.show()
```

#### Travessia em ordem simétrica

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import warnings

# Cria o grafo
G = nx.DiGraph()

# Adicionando arestas ao grafo
G.add_edges_from(arvore)

# Função de travessia em ordem simétrica
def inorder_traversal(G, node, visited=[]):
    neighbors = list(G.neighbors(node))
    
    # Se o nó tem vizinhos, visita o primeiro vizinho
    if neighbors:
        if neighbors[0] not in visited:
            inorder_traversal(G, neighbors[0], visited)
    
    # Visita o nó atual
    visited.append(node)

    # Visita os demais vizinhos
    for neighbor in neighbors[1:]:
        if neighbor not in visited:
            inorder_traversal(G, neighbor, visited)

    return visited

visited = inorder_traversal(G, arvore[0][0])

warnings.simplefilter("ignore")

# Exibe a lista de nós visitados
print(visited)

# Desenha o grafo
pos = nx.drawing.nx_pydot.graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, arrows=False)

# Exibe a figura
plt.show()
```

#### Travessia em ordem pós-ordem

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout
import warnings

# Cria o grafo
G = nx.DiGraph()

# Adicionando arestas ao grafo
G.add_edges_from(arvore)

# Função de travessia em pós-ordem
def postorder_traversal(G, node, visited=[]):
    for neighbor in G.neighbors(node):
        if neighbor not in visited:
            postorder_traversal(G, neighbor, visited)
    
    # Visita o nó atual após visitar todos os vizinhos
    visited.append(node)

    return visited

visited = postorder_traversal(G, arvore[0][0])

warnings.simplefilter("ignore")

# Exibe a lista de nós visitados
print(visited)

# Desenha o grafo
pos = nx.drawing.nx_pydot.graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, arrows=False)

# Exibe a figura
plt.show()
```
