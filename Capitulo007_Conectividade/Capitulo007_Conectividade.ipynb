<font color='black'><span style="font-size: 22px;"><strong>7.1 Conectividade em Grafos Não-Orientados</strong></span></font>

<hr style="border: 2px solid black;">

#### <font color='red'> Definição 1 </font>

<font color='red'> Um grafo não orientado é chamado **conexo** se houver um caminho entre cada par de vértices distintos do grafo. Um grafo não orientado que não está conexo é chamado de **desconexo**. Dizemos que **desconectamos** um grafo quando removemos vértices ou arestas, ou ambos, para produzir um subgrafo desconexo. (A remoção de um vértice implica na remoção de todas as arestas incidentes nele.)

#### <font color='blue'> Exemplo 1 </font>

<font color='blue'> O grafo $G$ da figura a seguir é conexo pois,  para cada par de vértices distintos há um caminho entre eles. No entanto, o grafo $H$ não o é. Por exemplo, não há caminho de $H$ entre os vértices $s$ e $v$.
 <center> <img src="fig7_1.jpg" alt="jpg_python" width="350" height=""> </center>

#### <font color='red'> Teorema 1 </font>

<font color='red'> Existe um caminho simples entre cada par de vértices distintos de um grafo não orientado conexo.

<font color='red'> Seja u e v dois vértices distintos do grafo não orientado conexo $G = (V, E)$. Como $G$ está conexo, existe pelo menos um caminho entre $u$ e $v$. Deixe $x_{0}, x_{1}, \ldots, x_{n}$, em que $x_{0} = u$ e $x_{n} = v$, seja a sequ\^encia de vértices de um caminho de menor comprimento. Esse caminho de menor comprimento é simples. Para ver isso, suponha que não seja simples. Então $ x_{i} = x_{j}$ para alguns $i$ e $j$ com $0 \leq i \leq j$. Isto significa que existe um caminho de $u$ para $v$ de menor comprimento com a sequ\^encia de vértices $x_{0}, x_{1}, \ldots, x_{i-1}, x_{j}, \ldots, x_{n}$ obtido pela exclusão das arestas correspondentes \`a sequ\^encia de vértices $x_{i}, \ldots, x_{j-1}$.

#### <font color='red'> Definição 2 </font>

<font color='red'> Um componente conexo de um grafo $G$ é um subgrafo conexo de $G$ que não é um subgrafo próprio de outro subgrafo conexo de $G$.

Decorre da **Definição 2** temos que um componente conexo de um grafo $G$ é um subgrafo conexo máximo de $G$. Um grafo $G$ não conexo possui dois ou mais componentes conexos que são separados e têm $G$ como sua união. 

#### <font color='blue'> Exemplo 2 </font>

<font color='blue'> Note que, no **Exemplo 1**, o grafo $H$ possui dois subgrafos conexos, formados respectivamente pelos conjuntos de vértices (e respectivas arestas) $H_{1} = \{s, t, x; st, sx\}$ e $H_{2} = \{u, v, w; uv, uw, vw\}$. Os subgrafos $H_{1}$ e $H_{2}$ são componentes conexas de $H$. Por outro lado, um subgrafo $H_{3}$ formado apenas pelos vértices $s$ e $t$ e pela aresta $\{s,t\}$ não é considerada uma componente conexa, embora seja conexo. Isso decorre do fato de existir o subgrafo conexo $H_{1}$ que o contém propriamente.


## <div style="margin-top: -20px;">

<font color='black'><span style="font-size: 18px;"><strong>7.1.1 Quão conexo é um Grafo?</strong></span></font>

<hr style="border: 2px solid black;">
</div>

Suponha que um grafo represente uma rede de computadores. Supondo que este grafo seja conexo nos podemos dizer que quaisquer dois computadores na rede podem se comunicar. No entanto, gostaríamos também de entender o quão confiável é essa rede. Por exemplo, ainda será possível que todos os computadores se comuniquem depois que um roteador ou um link de comunicação falhe? Para responder a esta e outras perguntas semelhantes apresentaremos alguns novos conceitos.

#### Vértices e arestas de corte

Às vezes, a remoção de um vértice de um grafo e de todas as bordas incidentes neste produz um subgrafo com mais componentes conexos que o original. Tais vértices são chamados de **vértices de corte**. Analogamente, uma aresta cuja remoção produz um grafo com mais componentes conexos do que no grafo original é chamada de **aresta de corte**. Observe que em um grafo representando uma rede de computadores, um vértice de corte e uma aresta de corte representam um roteador e um link essenciais que não podem falhar para que todos os computadores possam se comunicar.

#### <font color='blue'> Exemplo 3 </font>

<font color='blue'> Encontre os vértices de corte e as arestas de corte do grafo $G$ mostrado na figura a seguir. Em cada caso exiba os subgrafos conexos resultantes de cada remoção.<br><br>
 <center> <img src="fig7_2.jpg" alt="jpg_python" width="270" height=""> </center>

<font color='blue'> **Solução**: Temos dois vértices de corte, $b$ e $d$. O vértice $b$ quando removido produz as componentes conexas $G_{1} = \{a,c;ac\}$ e $G_{2} = \{d,e,f;de, df\}$; O vértice $d$ quando removido produz as componentes conexas $G_{3} = \{a,b,c,e;ab,ac,bc,be\}$ e $G_{4} = \{f\}$. Temos apenas uma aresta de corte a aresta $\{f,d\}$ cuja remoção produz as componentes conexas $G_{5} = \{a,b,c,d,e;ab,ac,bc,bd,be,de\}$ e $G_{6} = \{f\}$.

#### Grafos completos

Nem todos os grafos possuem vértices de corte. Por exemplo, o grafo completo $K_{n}$, com $n \geq 3$, não possui vértices de corte. Quando se remove um vértice de $K_{n}$ e todas as bordas incidentes a ele, o subgrafo resultante é o grafo completo $K_{n-1}$, um grafo conexo. Grafos conexos sem vértices de corte são chamados de **grafos não separáveis** e, além disso são considerados mais conexos do que aqueles com um vértice de corte. Podemos estender essa noção definindo uma medida mais precisa de conectividade de grafos baseada no número mínimo de vértices que podem ser removidos para desconectar um grafo.

#### <font color='red'> Definição 3 </font>

<font color='red'> Um subconjunto $V'$ do conjunto de vértices $V$ de $G = (V, E)$ é um **corte de vértices**, ou **conjunto de separação**, se $G - V'$ for desconexo. Define-se a conectividade de vértices de um grafo não completo $G$, denotada por $\kappa(G)$, como sendo o número mínimo de vértices em um corte de vértices.

#### <font color='red'> Definição 4 </font>

<font color='red'> Um subconjunto $E'$ do conjunto de arestas $E$ de $G = (V,E)$ é um **corte de arestas** se $G-E'$ for desconexo. Define-se a conectividade de arestas de um grafo $G$, denotada por $\lambda(G)$ como sendo o número mínimo e arestas em um corte de arestas. 

Note que a **Definição 4** nos diz que todo grafo conexo com mais de um vértice pode ser desconectado removendo-se todas as arestas incidentes em um dos seus vértices. Além disso, se $\lambda(G) = 0$ então, $G$ é não conexo ou $G$ possui apenas um vértice. Isso define bem $\lambda$.

#### <font color='red'> Teorema 2 </font>

<font color='red'> Segue que se $G$ é um grafo com $n$ vértices, então $0 \leq \lambda(G) \leq n-1$.

#### <font color='blue'> Exemplo 4 </font>

<font color='blue'> Encontre a conectividade de vértices e por arestas de cada um dos grafos na figura a seguir:<br><br>
 <center> <img src="fig7_3.jpg" alt="jpg_python" width="500" height=""> </center>

<font color='blue'> **Solução**: *Grafo $G$:* Note que o vértice $b$ de $G$ é um vértice de corte donde segue que $\kappa(G) = 1$. De fato, removendo-se $b$ produzimos subgrafos conexos $G_{1} = \{a,c;ac\}$ e $G_{2} = \{d,e,f;de,df,ef\}$. Ainda considerando o grafo $G$, temos $\lambda(G) = 2$, pois removendo-se duas arestas, digamos $\{b,d\}$ e $\{b,e\}$ produzimos dois subgrafos conexos $G_{3} = \{a,b,c;ab,ac,bc\}$ e $G_{4} = \{d,e,f;de,df,ef\}$;<br>
 *Grafo $H$:* Considerando o grafo $H$, temos $\kappa(H) = 2$ pois, removendo-se os vértices $b$ e $f$ produzimos dois subgrafos conexos $H_{1} = \{a,e;ae\}$ e $H_{2} = \{c,d,g,h;cd,cg,gh\}$. Além disso, temos que $\lambda(H) = 3$ pois, precisamos remover ao menos três arestas para produzir subgrafos desconexos. Em particular, se removermos as arestas $\{a,b\}$, $\{a,e\}$ e $\{a,f\}$ produziremos os subgrafos conexos $H_{3} = \{a\}$ e $H_{4} = \{b,c,d,e,f,g,h;be,bc,bd,bh,cd,cg,df,fe,fg,fh,gh\}$.

#### Relação entre graus dos vértices e a conectividade de um grafo

Quando $G = (V, E)$ é um grafo conexo incompleto com pelo menos trê vértices, o grau mínimo de um vértice de $G$ é um limite superior para $\kappa(G)$ e $\lambda(G)$. Ou seja, $\kappa(G) \leq min_{v \in V} \, deg(v)$ e $\lambda(G) \leq min_{v\in V} \, deg(v)$. Temos, a seguinte inequação:<br><br>
<font color='red'>$$\kappa(G) \leq \lambda(G) \leq min_{v \in V}\, deg(v).$$

As definições a seguir fundamentam a noção de concetividade em grafos orientados:

#### <font color='red'> Definição 5 </font>

<font color='red'> Um grafo orientado é **fortemente conexo** se houver um caminho de $a$ para $b$ e de $b$ para $a$ sempre que $a$ e $b$ forem vértices no grafo.

#### <font color='red'> Definição 6 </font>

<font color='red'> Um grafo orientado é **fracamente conexo** se houver um caminho entre cada dois vértices no grafo não orientado associado.

#### <font color='blue'> Exemplo 5 </font>

<font color='blue'> Os grafos orientados $G$ e $H$ mostrados a seguir estão fortemente conexos? Eles estão fracamente conexos?<br><br>
 <center> <img src="fig7_4.jpg" alt="jpg_python" width="300" height=""> </center>

<font color='blue'> **Solução**: $G$ é fortemente conexo pois existe um caminho entre quaisquer dois vértices neste grafo orientado (o leitor deve verificar isso). Portanto, $G$ também é fracamente conexo. O grafo $H$ não é fortemente conexo. Não há caminho orientado de $s$ para $t$ neste grafo. No entanto, $H$ é fracamente conexo, pois existe um caminho entre quaisquer dois vértices no grafo não orientado associado a $H$ (o leitor deve verificar isso).

#### Componentes fortes

Os subgrafos de um grafo orientado $G$ que estão fortemente conexos, mas não contidos em subgrafos maiores fortemente conexos são chamados de **componentes fortemente conexos** ou **componentes fortes** de $G$. Note que se $a$ e $b$ são dois vértices em um grafo direcionado, seus componentes fortes são os mesmos ou são disjuntos.

O grafo $H$ apresentado no **Exemplo 5** possui três componentes fortemente conexos, $H_{1}$ consistindo no vértice $s$; $H_{2}$ consistindo no vértice $w$; e o subgrafo $H_{3}$ consistindo dos vértices $t$, $u$, e $v$ e arestas $(t, u)$, $(u, v)$ e $(v, t)$. Como dito anteriormente, $H_{i} \cap H_{j} = \emptyset$ sempre que $i \neq j$ com $i,j \in \{1,2,3\}$.

##### Exibindo a conectividade de vértices, o conjunto mínimo de vértices de corte, o esboço do grafo original e o esboço das componentes conexas resultantes

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Crie o grafo G e adicione arestas conforme necessário
G = nx.Graph()
G.add_edges_from(grafo_G)  # Substitua 'grafo_G' para 'grafo_H' ('grafo_I' >>> erro)


# Encontre conjuntos de corte
cut_sets = list(nx.all_node_cuts(G))

# Encontre componentes conexas após a remoção dos vértices de corte
components = []
for cut_set in cut_sets:
    temp_G = G.copy()
    temp_G.remove_nodes_from(cut_set)
    components.extend(list(nx.connected_components(temp_G)))

# Crie um novo grafo com as componentes conexas
new_G = nx.Graph()
for comp_nodes in components:
    new_G.add_nodes_from(comp_nodes)
    for node1 in comp_nodes:
        for node2 in comp_nodes:
            if G.has_edge(node1, node2):
                new_G.add_edge(node1, node2)

# Configuração do layout de subplots
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(1, 2, width_ratios=[1, 1.5])

# Plotagem do grafo original
ax1 = plt.subplot(gs[0])
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, font_size=10, node_color='lightgray')
nx.draw_networkx_nodes(G, pos, nodelist=cut_sets[0], node_color='red', node_size=1000)
ax1.set_title("Grafo Original")

# Plotagem do novo grafo com componentes
ax2 = plt.subplot(gs[1])
pos_new = nx.spring_layout(new_G)
nx.draw(new_G, pos_new, with_labels=True, node_size=1000, font_size=10, node_color='lightgray')
nx.draw_networkx_edges(new_G, pos_new, edgelist=new_G.edges(), edge_color='black')
ax2.set_title("Componentes após Remoção")

plt.tight_layout()
plt.show()

print("Conjuntos de corte:", cut_sets)
print("Componentes conexas após remoção dos vértices de corte:")

for idx, comp in enumerate(components, start=1):
    subgraph = G.subgraph(comp)
    print(f"Componente {idx}:")
    print("  Vértices:", comp)
    print("  Arestas:", subgraph.edges())
    print()
```

##### Exibindo a conectividade de arestas, o conjunto mínimo de arestas de corte, o esboço do grafo original e o esboço das componentes conexas resultantes

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Crie o grafo G e adicione arestas conforme necessário
G = nx.Graph()
G.add_edges_from(grafo_G) # Substitua o 'grafo_G' pelo 'grafo_H' ('grafo_I' >>> erro)

# Encontre conjuntos mínimos de cortes de aresta
cut_set = nx.minimum_edge_cut(G)

# Encontre componentes conexas após a remoção das arestas de corte
temp_G = G.copy()
for edge in cut_set:
    temp_G.remove_edge(*edge)

components = list(nx.connected_components(temp_G))

# Crie um novo grafo com as componentes conexas
new_G = nx.Graph()
for comp_nodes in components:
    new_G.add_nodes_from(comp_nodes)
    new_G.add_edges_from(G.subgraph(comp_nodes).edges())

# Configuração do layout de subplots
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(1, 2, width_ratios=[1, 1.5])

# Plotagem do grafo original
ax1 = plt.subplot(gs[0])
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, font_size=10, node_color='lightgray')
nx.draw_networkx_edges(G, pos, edgelist=cut_set, edge_color='red')
ax1.set_title("Grafo Original")

# Plotagem do novo grafo com componentes
ax2 = plt.subplot(gs[1])
pos_new = nx.spring_layout(new_G)
nx.draw(new_G, pos_new, with_labels=True, node_size=1000, font_size=10, node_color='lightgray')
nx.draw_networkx_edges(new_G, pos_new, edgelist=new_G.edges(), edge_color='black')
ax2.set_title("Componentes após remoção de arestas")

plt.tight_layout()
plt.show()

print("Conjuntos de corte:", cut_set)
print("Componentes conexas após remoção das arestas de corte:")

for idx, comp in enumerate(components, start=1):
    subgraph = G.subgraph(comp)
    print(f"Componente {idx}:")
    print("  Vértices:", comp)
    print("  Arestas:", subgraph.edges())
    print()
```

##### Verificando se um grafo é fortemente conexo, fracamente conexo ou nenhum dos dois casos

```python
import networkx as nx

# Crie um grafo orientado de exemplo
grafo_orientado = nx.DiGraph()
grafo_orientado.add_edges_from(grafo_G) # Substitua 'grafo_G' por 'grafo_H' ou 'grafo_I'

# Verifica se o grafo é fortemente conexo
if nx.is_strongly_connected(grafo_orientado):
    print("O grafo é fortemente conexo.")
else:
    if nx.is_weakly_connected(grafo_orientado):
        print("O grafo é fracamente conexo apenas.")
    else:
        print("O grafo não é nem ao menos fracamente conexo.")
```

##### Exibindo as componentes fortemente conectadas de um grafo

```python
import networkx as nx

# Crie um grafo orientado de exemplo
grafo_orientado = nx.DiGraph()
grafo_orientado.add_edges_from(grafo_G)

# Encontra as componentes fortemente conectadas
componentes_fortemente_conectadas = list(nx.strongly_connected_components(grafo_orientado))

# Exibe as informações das componentes fortemente conectadas
for idx, componente in enumerate(componentes_fortemente_conectadas):
    print(f"Componente {idx + 1}, vértices: {componente}, arestas: {grafo_orientado.subgraph(componente).edges}")
```

##### Exibindo as componentes fortemente conectadas de um grafo juntamente com um esboço

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm

grafo_orientado = nx.DiGraph()
grafo_orientado.add_edges_from(grafo_I)

# Encontra as componentes fortemente conectadas
componentes_fortemente_conectadas = list(nx.strongly_connected_components(grafo_orientado))

# Prepara o layout do grafo original
pos = nx.spring_layout(grafo_orientado, seed=42)

# Plotagem do grafo original
plt.figure(figsize=(10, 8))
nx.draw(grafo_orientado, pos, with_labels=True, node_size=1000, font_size=10, node_color='lightgray', edge_color='gray')

# Gera uma paleta de cores que se adapta ao número de componentes fortemente conectadas
num_cores = len(componentes_fortemente_conectadas)
colors = cm.rainbow([i/num_cores for i in range(num_cores)])

patches = []  # Para guardar os patches da legenda
for idx, (componente, color) in enumerate(zip(componentes_fortemente_conectadas, colors)):
    subgraph = grafo_orientado.subgraph(componente)
    nx.draw(subgraph, pos, with_labels=True, node_size=1000, font_size=10, node_color=color, edge_color=color, linewidths=2)
    
    # Imprime as informações das componentes
    vertices_str = ', '.join(map(str, componente))
    arestas_str = ', '.join(map(str, subgraph.edges))
    print(f"Componente {idx + 1}, vértices: {{{vertices_str}}}, arestas: [{arestas_str}]")
    
    # Adiciona um patch para a legenda
    patches.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=f"Componente {idx + 1}"))

# Legenda para as componentes
plt.legend(handles=patches, loc='best')

plt.show()
```
